# Fazendo análise de notas aleatórias

```python
from random import randrange, seed

import matplotlib.pyplot as plt
```

```python
seed(5)

notas = []

for nota in range(8):
  notas.append(randrange(0,11))

notas
```

```python
x = list(range(1, 9))
y = notas

plt.plot(x, y, marker = 'o')
plt.title('Notas de Matemática')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
print(notas)
```

```python
seed(6)

notas = []

for nota in range(8):
  notas.append(randrange(0,11))
```

```python
x = list(range(1, 9))
y = notas

plt.plot(x, y, marker = 'o')
plt.title('Notas de Matemática')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
print('-> NOTAS DAS AVALIAÇÕES PARCIAIS:')
for nota in range(len(notas)):
  if notas[nota]>=7:
    print(f'\t\033[34mAP {nota+1}: {notas[nota]}\033[m')
  else:
    print(f'\t\033[31mAP {nota+1}: {notas[nota]}\033[m')
```
